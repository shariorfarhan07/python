```
class contact:
  name='null'
  address='null'
  email='null'
  phone='null'
  
  def details(self):
    print('name',self.name)
    print('address',self.address)
    print('email',self.email)
    print('phone',self.phone)
  

c1=contact()#c1 variable contact() constructor
c2=contact()#c2=contact() object create 
c3=contact()
c4=contact()
c1.name="farhan"
c1.address='shantinogor'
c1.email="sh557ariorfarhan"
c1.phone='01749577'
c2.name="sabrina"
c2.address='bonani'
c2.email="sab@gmail.com"
c2.phone='0174945626'
c3.name="rohim"
c3.address='mohakhali'
c3.email="rohomi@gmail.com"
c3.phone='659519'
print('---------')
c1.details()
print('---------')
c2.details()
print('---------')
c3.details()
print('---------')
c4.details()
print('---------')
```
