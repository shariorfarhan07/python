```
import random
random.seed(300)

def makegraph(noOfterm,branch,max,min,n):
  if n==None:
    n=node()
  if noOfterm!=1:
    for i in range(branch):
      n.addchild(node())
      makegraph(noOfterm-1,branch,max,min,n.child[i])
    
  else:
    for i in range(branch):
      c=node()
      c.value=random.randint(min,max);
      n.addchild(c)
  
  return n









a=makegraph(noOfterm,branch,max,min,None)



```

```
def max(a,b):
  if a>b:
    return a
  return b
def min(a,b):
  if a<b:
    return a
  return b
```

```
import math as m
count=0
def minimax(node, depth, maximizing):
  global count
  if depth == 0:
    count+=1
    return node.value 
  
  if maximizing:
    maxEval = -m.inf
    for child in node.child:
      eval = minimax(child, depth - 1, False)
      maxEval  =maxEval if maxEval>eval else eval 
    return maxEval
  else:
    minEval = m.inf
    for child in node.child:
      eval = minimax(child, depth - 1, True)
      minEval =minEval if minEval<eval else eval 
     
    return minEval
```

```
import math as m
count=0
def minimaxAB(node, depth,alpha, beta, maximizing):
  global count
  if depth == 0:
    count+=1
    return node.value 
  
  if maximizing:
    maxEval = -m.inf
    for child in node.child:
      eval = minimaxAB(child, depth - 1,alpha, beta, False)
      maxEval =maxEval if maxEval>eval else eval     
      alpha =alpha if alpha>eval else eval             
      if beta <= alpha:
        break
    return maxEval
  else:
    minEval = m.inf
    for child in node.child:
      eval = minimaxAB(child, depth - 1,alpha, beta, True)
      minEval =minEval if minEval<eval else eval     
      beta =beta if beta<eval else eval
      if beta <= alpha:
        break     
    return minEval
```
