```
def math(a,b,s):
  if s=="add":
    return a+b
  elif s=="divd":
    return a/b
  elif s=="sub":
    return a-b
  elif s=="multi":
    return a*b
  else :
    return "wrong keyword"
  


print(math(2,10,"sub"))
print(math(2,10,"ad"))
print(math(2,10,"add"))
print(math(2,10,"divd"))
```

```
a1=[1,5,9,8,7,6,3,2]
a2=[551,555,95,558,557,655,355,255]
a3=[155,555,955,855,755,655,355,255]
a4=[1000,5000,9000,8000,7000,600,300,2000]
a5=[10,50,90,80,70,60,30,20]

def maxlist(a):
  max=a[0]
  for i in a:
    if max<i:
      max=i
  return max
maxl=[maxlist(a1),maxlist(a2),maxlist(a3),maxlist(a4),maxlist(a5)]
print(maxl)
print(maxlist(maxl))
```
