```python
#palidrome problem circular array
s=[20,10,0,0,0,10,20,30]
n=[10,20,0,0,0,10,20,30]
def checkpalindrome(s,start,size):
    end=(start+size-1)%len(s)
    for i in range(size):
      if s[start]!=s[end]:
        return False
      start+=1
      end-=1
      if start == len(s):
        start=0
      if end==-1:
        end=len(s)-1
    return True
print(checkpalindrome(s,5,5))
print(checkpalindrome(n,5,5))
```

```python
#6. Array series
def pattern(n):
  out=[]
  s=[]
  for i in range(n,0,-1):
    s.append(i)
  out=s+out
  for i in range(n-1):
    s[i]=0
    out=s+out
  return out

pattern(4)
```

```python
#7. Max Bunch Count
def MaxBunchCount(a):
  count=0
  for i in range(len(a)-1):
    if a[i]==a[i+1]:
      count+=1
  return count
a=[1, 2, 2, 3, 4, 4, 4]
MaxBunchCount(a)
```

```python
def removeAll(s,size,i):
  b=i
  for i in range(size):
    try:
      s.remove(b)
      s.append(0)
    except:
      pass
  return s
   

source=[10,2,30,2,50,2,2,60,0,0]
removeAll(source,8,2)
```
