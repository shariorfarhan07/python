```python
class node:
  def __init__(self,v,n=None):
    self.value=v
    self.next=n



class linklist:
  def __init__(self):
    self.head=None
    self.size=0

  def append(self,n):
    if self.head==None:
      self.head=node(n)
    else:
      tail=self.head
      while tail.next!=None:
        tail=tail.next
      tail.next=node(n)
    self.size+=1


  def printlist(self):
    tail=self.head
    while tail!=None:
      print(tail.value,end='->')
      tail=tail.next
    

  def count(self):
    tail=self.head
    count=0
    while tail!=None:
      count+=1
      tail=tail.next
    return count



  def insert(self,value,ind):
    tail=self.head
    if ind<0 or ind>self.size:
      print('invalid index')
    elif ind==0:
      self.head=node(value,tail)
    elif ind==self.size:
      while tail.next!=None:
        tail=tail.next
      tail.next=node(value)
      self.size+=1       
    else:
      for i in range(ind-1):
        tail=tail.next
      r=tail.next.next
      tail.next=node(value,r)
      self.size+=1
      


    
  def delete(self,ind):
    if ind<0 or ind>self.size:    
      print('invalid index')
    elif ind==0:
      self.head=self.head.next        
    else:
      tail=self.head
      for i in range(ind-1):
        tail=tail.next
      tail.next=tail.next.next
      
    
```

```python
l=linklist()
l.append(1)
l.append(2)
l.append(3)
l.append(4)
l.append(5)
l.append(6)
l.printlist()
print(l.size)  
print(l.count())
l.insert(100,0) 
l.insert(1000,6) 
l.printlist()
l.delete(1)
l.delete(6)
l.printlist()
```
