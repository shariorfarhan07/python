```python
def binaryFloat(a,n,c):
  a=a*2
  c+=1
  if n>=c:
    return str(int(a))+binaryFloat(a-int(a),n,c)
  return ""

def binary(a):
  f=a/2
  a=a//2
  if a!=0 or f !=0:
    if f-a>0:
      return binary(a)+"1"
    else:
      return binary(a)+"0"
  return ""
  
def insert (source_str, insert_str, pos):
  return source_str[:pos]+insert_str+source_str[pos:]

def bitHex(b):
  b=int(b,2)
  #print(b)
  if b>=10:
    return chr(ord('A')+(b-10))
  else:
    return str(b)
 

def hexx(a):
  if len(a)==0:
    return ''
  if len(a)!=0:
    return bitHex(a[:4])+hexx(a[4:])

def toint(a):
  l=a.find('.')
  first=a[:l]
  first=first[::-1]
  last=a[l+1:]
  d=0
  c=0
  for i in first:
    d+=int(i)*2**c
    c+=1
  c=1
  for i in last:
    d+=int(i)*2**-c
    c+=1
  return d
```
