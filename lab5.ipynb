```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("heart disease classification dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(df[['chol']])
df[['chol']] = impute.transform(df[['chol']])



```

```python
impute.fit(df[['trestbps']])
df[['trestbps']] = impute.transform(df[['trestbps']])
df.isnull().sum()
```

```python
impute.fit(df[['thalach']])
df[['thalach']] = impute.transform(df[['thalach']])
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Set up the LabelEncoder object
enc = LabelEncoder()

# Apply the encoding to the "Accessible" column
df['sex'] = enc.fit_transform(df['sex'])
df['target'] = enc.fit_transform(df['target'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df)
```

```python
# transform data
df_scaled = scaler.transform(df)
df_scaled
```

```python
df.columns
```

```python
df=pd.DataFrame(df_scaled,columns=df.columns)
```
