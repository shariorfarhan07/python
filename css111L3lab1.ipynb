```
#harry, hermione

def matching(a,b):
  a=list(a)
  b=list(b)
  g=''
  for i in range(min(len(a),len(b))):
    if a[i] in b:
      g+=a[i]
      a[i]='~'
  return g

a=input().split(',')
o=matching(a[0],a[1])+matching(a[1],a[0])
print(o)


```
