```python
class node:
  def __init__(self,v,n=None):
    self.value=v
    self.next=n

class stack:
  def __init__(self):
    self.head=None
    self.size=0
  
  def printlist(self):
   tail=self.head
   while tail!=None:
     print(tail.value,end='->')
     tail=tail.next
   print('None')

  def push(self,a):
    self.head=node(a,self.head)
    self.size+=1
    return True


  def pop(self):
    a=self.head.value
    self.head=self.head.next
    self.size-=1
    return a.value





  def pick(self):
    return self.head.value

l=stack()
l.push(1)
l.push(2)
l.push(3)
l.push(4)
l.printlist()
l.pop()
l.push(40)
l.printlist()
l.pop()
l.printlist()
l.pick()
```

Enqueue: Adds an item to the queue. If the queue is full, then it is said to be an Overflow condition – Time Complexity : O(1)
Dequeue: Removes an item from the queue. The items are popped in the same order in which they are pushed. If the queue is empty, then it is said to be an Underflow condition – Time Complexity : O(1)
Front: Get the front item from queue – Time Complexity : O(1)

```python
class node:
  def __init__(self,v,n=None):
    self.value=v
    self.next=n
#Enqueue
#Dequeue
#Front
#Rear
class queue:
  def __init__(self):
    self.head=None
    self.size=0

  def printlist(self):
   tail=self.head
   while tail!=None:
     print(tail.value,end='->')
     tail=tail.next
   print('None')
  
  def enqueue(self,a): # append 
    tail=self.head
    if tail!=None:
      while tail.next!=None:
        tail=tail.next
      tail.next=node(a)
    else:
      self.head=node(a)
    self.size+=1



  def dequeue(self):
    b=self.head
    if self.head!=None:
      self.head=self.head.next
    return b.value



   
    
  def front(self,a):
    return self.head.value
  def rear(self,a):
    tail=self.head
    while tail.next!=None:
      tail=tail.next
    return tail.value




l=queue()
l.enqueue(1)
l.enqueue(2)
l.enqueue(3)
l.enqueue(4)
l.enqueue(5)
l.dequeue()
l.printlist()
  
```
