```
def add(*a):
  #a=(5,5,5,5)
  sum=0
  for i in a:
    sum=i+sum
  print(a,'=',sum)

  
 
add(5,5,5,5)
add(1,5,9,8,7)
add(5,6,5,4,6,7)
```
