```
a=[1,2,3,4,'farhan',5,85,5,5,6,7,8,9]
#Python List insert()
#insert an element to the list
a.insert(4,5000)
print(a)
#Python List pop()
#Removes element at the given index
print(a.pop())
print(a)
print(a.pop())
print(a)
#Python List remove()
#Removes item from the list
print(a.remove('farhan'))
print(a)
print(a.remove(5))
print(a)
#print(a.remove(5555))
#print(a)

#Python List reverse()
#reverses the list
a.reverse()
print(a)
#Python List sort()
a.sort()
print(a)
a.sort(reverse=True)
print(a)
#sorts elements of a list
```
