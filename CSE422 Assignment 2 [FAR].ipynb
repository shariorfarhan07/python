```python
import math
import random
import numpy as np

cost_map = [[math.inf, 2, 4, math.inf, math.inf, math.inf],
            [2, math.inf, 3, 5, math.inf, 15],
            [4, 3, math.inf, 3, 6, 11],
            [math.inf, 5, 3, math.inf, 1, 6],
            [math.inf, math.inf, 6, 1, math.inf, 4],
            [math.inf, 15, 11, 6, 4, math.inf]]

path = [[1, 2],
        [0, 2, 3, 5],
        [0, 1, 3, 4, 5],
        [1, 2, 4, 5],
        [2, 3, 5],
        [1, 2, 3, 4]]

def fitness_population(population):
    return [fitness(chromosome) for chromosome in population]

def fitness(chromosome):
    path_cost = 0

    for i in range(len(chromosome)):
        if chromosome[i] != None:
            path_cost += cost_map[i][chromosome[i]]

    return path_cost

def select(population, fitness, number):
    p = []
    f = []
    for chromosome in population:
        c = []
        for gene in chromosome:
            c.append(gene)
        p.append(c)
    
    for fit in fitness:
        f.append(fit)
    
    selector = []
    for i in range(len(p)):
        selector.append([f[i], p[i]])

    for i in range(len(selector)):
        max_idx = i
        for j in range(i, len(selector)):
            if selector[j][0] > selector[max_idx][0]:
                max_idx = j
        if i != max_idx:
            temp = selector[i]
            selector[i] = selector[max_idx]
            selector[max_idx] = temp
    
    return selector[:number]

def crossover(x, y):
    child = []
    pivot = random.randint(0, len(x) - 1)
    
    for i in range(pivot):
        child.append(x[i])
    
    for i in range(pivot, len(y)):
        child.append(y[i])
    
    return child

def mutate(child):
    idx = random.randint(0, len(child) - 1)
    val = random.randint(0, len(child) - 1)

    child[idx] = val

def fill(chromosome):
    city = [0, 1, 2, 3, 4, 5, None]

    for gene in chromosome:
        city.remove(gene)
    
    idx = 0
    for i in range(len(chromosome)):
        if chromosome[i] == None:
            idx = i
    
    chromosome[idx] = city[0]

def generate_chromosome():
    chromosome = [None, None, None, None, None, None]
    connection = []

    while len(connection) < 6:
        node = random.randint(0, 5)
        while node in connection:
            node = random.randint(0, 5)
        connection.append(node)

    source = connection[0]
    for i in range(1, len(connection) - 1):
        chromosome[source] = connection[i]
        source = connection[i]

    chromosome[connection[-1]] = connection[0]

    fill(chromosome)

    if fitness(chromosome) == math.inf:
        return generate_chromosome()

    # fill(chromosome)
    return chromosome

user_input = None

population = []

for i in range(3):
    population.append(generate_chromosome())

print('Answer for 2.1')
print('=' * 14)
for chromosome in population:
    print(chromosome)


print('\n\nAnswer for 2.2')
print('=' * 14)
print('i.   Gene are each elements of the array. For example, 1, 3, 0, etc.')
print('ii.  The collection of gene is known as chromosome. Therefore, each array represents a chromosome.')


print('\n\nAnswer for 2.3')
print('=' * 14)
print('Explain the above fitness function in your own words.')


print('\n\nAnswer for 2.4')
print('=' * 14)
fitnesses = fitness_population(population)
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
print('|%20s|%10s|' % ('Chromosome', 'Fitness'))
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
for chromosome, fit_value in zip(population, fitnesses):
    print('|%20s|%10s|' % (chromosome, fit_value))
    print('+%20s+%10s+' % ('-' * 20, '-' * 10))

print('\n\nSelected chromosomes:')
selected_chromosome = select(population, fitnesses, 2)

for chromosome in selected_chromosome:
    print('   ', chromosome[1])


print('\n\nAnswer for 2.5')
print('=' * 14)
child = crossover(selected_chromosome[0][1], selected_chromosome[1][1])
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
print('|%20s|%10s|' % ('Child', 'Fitness'))
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
print('|%20s|%10s|' % (child, fitness(child)))
print('+%20s+%10s+' % ('-' * 20, '-' * 10))


print('\n\nAnswer for 2.6')
print('=' * 14)
mutate(child)
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
print('|%20s|%10s|' % ('Child', 'Fitness'))
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
print('|%20s|%10s|' % (child, fitness(child)))
print('+%20s+%10s+' % ('-' * 20, '-' * 10))
```
