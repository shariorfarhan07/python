```python
def calc(a,b,op):
  try:
    a=int(a)
    b=int(b)
  except:
    a=float(a)
    b=float(b)

  if '/' ==op or '\\' ==op:
    return a/b
  elif '+' ==op:
    return a+b
  elif '-' ==op:
    return a-b
  elif '%' ==op:
    return a%b
  elif '*' ==op:
    return a*b
  
```

```python
postfix="4 + 3 * ( 6 * 3 - 12 )"
operators='\ * - + %'.split(' ')
stack=[]
display=[]
postfix=postfix.split(' ')
pointer_at=0
#insert number in stack 
for j,i in enumerate(postfix):
  if i not in operators:
    stack.append(i)
    display.append(i)
    pointer_at=j
    print(postfix[j:],'\npointer index-->', pointer_at,'\nbottom-->',stack,'--> stack top\n\n')
    #pull two number from stack and calculate using operator and put the calculated value back in stack 
  else:
    print(postfix[j:],'\npointer index-->', pointer_at,'\nbottom-->',stack,'--> stack top\n\n')
    number1=stack.pop()
    number2=stack.pop()
    op=i
    value=calc(number2,number1,op)
    display.append(value)
    print(number2,op,number1,'=',value)
    
   
    stack.append(str(value))



```

```python
class node:
  def __init__(self,v=None,n=None,p=None):
    self.value=v
    self.next=n
    self.prev=p



def listTransfrom(head):
  tail=head
  while tail.next!=None:
    tail.next.prev=tail
    tail=tail.next
  return head




  
head=node(1,node(2,node(3)))
listTransfrom(head)
```
