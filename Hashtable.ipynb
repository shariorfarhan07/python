```python
class hastable:
  def __init__(self,a):
    self.k=[-1]*len(a)
    self.d={'a':1,'b':2,'c':3,'d':4,'e':5,'f':6,'g':7,'h':8,'i':9,'j':10,'k':11,'l':12,'m':13,'n':14,'o':15,'p':16,'q':17,'r':18,'s':19,'t':20,'u':21,'v':22,'w':23,'x':24,'y':25,'z':26,'1':27,'2':28,'3':29,'4':30,'5':31,'6':32,'7':33,'8':34,'9':35}
    for value in a:
      sum=0
      for char in value:
        sum=self.d[char.lower()]
      ind=sum%len(self.k)
      if self.k[ind]==-1:
        self.k[ind]=value
      else:
        while self.k[ind]!=-1:
          ind+=1
          ind=ind%len(self.k)
        self.k[ind]=value
 
    

test= hastable(['apple','abc','def','ghk','lmnop23'])
print(test.k)
      
#task 2

    
```
