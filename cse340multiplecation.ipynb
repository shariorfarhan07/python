```python
def hex_add(val, incre):
    return hex(int(str(val), 16) + incre)[2:].upper()

def binf(val, bit):
    if type(val) == type(0):
        val = bin(val)[2:]
    return '0' * abs(len(val) - bit) + str(val)

def twos_complement(val, bit):
    assert val < 0

    val *= -1
    val = binf(bin(val)[2:], bit)
    complement = ['1', '0']
    for i in val:
        val += complement[int(i)]
        val = val[1:]

    # print(val)
    val = bin(int(val, 2) + 1)[2:]
    val = val[abs(len(val) - bit):]
    
    return val

def multiply(multiplier, multiplicand):
    product = '0' * len(multiplicand)
    # print('Multiplicand   :', int(multiplicand, 2))
    # print('Multiplier     :', int(multiplier, 2))
    print()
    print('Expected Result:', int(multiplier, 2) * int(multiplicand, 2))
    print()

    iteration = 0

    print("+%20s+%20s+%20s+%20s+" % ('-' * 20, '-' * 20, '-' * 20, '-' * 20))
    print("|%20s|%20s|%20s|%20s|" % ('Iteration', 'Multiplier', 'Multiplicand', 'Product'))
    print("+%20s+%20s+%20s+%20s+" % ('-' * 20, '-' * 20, '-' * 20, '-' * 20))

    print("|%20s|%20s|%20s|%20s|" % (iteration, multiplier, multiplicand, product))
    print("+%20s+%20s+%20s+%20s+" % ('-' * 20, '-' * 20, '-' * 20, '-' * 20))
    for iteration in range(1, len(multiplier) + 1):
        iteration_str = str(iteration)
        if multiplier[-1] == '1':
            product = bin(int(product, 2) + int(multiplicand, 2))[2:]
            product = '0' * abs(len(multiplicand) - len(product)) + product
            print("|%20s|%20s|%20s|%20s|" % (iteration_str, multiplier, multiplicand, product))
            iteration_str = ''

        multiplicand = multiplicand[1:] + '0'
        print("|%20s|%20s|%20s|%20s|" % (iteration_str, multiplier, multiplicand, product))
        multiplier = '0' + multiplier[:-1]
        print("|%20s|%20s|%20s|%20s|" % ("", multiplier, multiplicand, product))
        print("+%20s+%20s+%20s+%20s+" % ('-' * 20, '-' * 20, '-' * 20, '-' * 20))

    print('Result:', product, '(', int(product, 2), ')')
    return product


student_id = input("Enter your ID: ")
# student_id = '18361264'
id = [0] + [int(i) for i in list(student_id)]

reg = {'t1': id[5], 't2': id[6], 's1': id[7], 's5': id[8], 's6': 11}
t1 = 't1'
t2 = 't2'
s1 = 's1'
s5 = 's5'
s6 = 's6'


print('=' * 80)
print('ANSWER FOR QUESTION 1')
print('=' * 80, end = '\n\n\n')


print('Corresponding register numbers:')
for key, data in reg.items():
    print(key, '=', data)

print('First memory address:')
memory_address = int(str(id[-4]) + str(id[-3]) + str(id[-2]) + str(id[-1])) * 4
print(memory_address)

instructions = ['Loop: sll  $t2, $s1, 2   ',
                '      add  $t2, $t2, $s6 ',
                '      lw   $t1, 0($t2)   ',
                '      bne  $t1, $s5, Exit',
                '      addi $s1, $s1, 2   ',
                '      addi $s1, $s1, -1  ',
                '      j    Loop          ',
                'Exit: …                  ']

print("+%25s+%15s+%41s+" % ('-' * 25, '-' * 15, '-' * 41))
print("|%25s|%15s|%41s|" % ('Instruction', 'Memory Address', 'Machine Code'))
print("+%25s+%15s+%6s+%6s+%6s+%6s+%6s+%6s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6))

# print("|%25s|%15s|%6s|%6s|%6s|%6s|%6s|%6s|" % (instructions[0], hex_add(memory_address, 4 * 0), binf(0, 6), binf(0, 5), binf(reg[s1], 5), binf(reg[t2], 5), binf(2, 5), binf(0, 6)))
print("|%25s|%15s|%6s|%6s|%6s|%6s|%6s|%6s|" % (instructions[0], memory_address + 4 * 0, 0, 0, reg[s1], reg[t2], 2, 0))
print("+%25s+%15s+%6s+%6s+%6s+%6s+%6s+%6s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6))

# print("|%25s|%15s|%6s|%6s|%6s|%6s|%6s|%6s|" % (instructions[1], hex_add(memory_address, 4 * 1), binf(0, 6), binf(reg[t2], 5), binf(reg[s6], 5), binf(reg[t2], 5), binf(0, 5), binf(32, 6)))
print("|%25s|%15s|%6s|%6s|%6s|%6s|%6s|%6s|" % (instructions[1], (memory_address + 4 * 1), 0, reg[t2], reg[s6], reg[t2], 0, 32))
print("+%25s+%15s+%6s+%6s+%6s+%6s+%6s+%6s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6, '-' * 6))

# print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[2], hex_add(memory_address, 4 * 2), binf(35, 6), binf(reg[t2], 5), binf(reg[t1], 5), binf(0, 16)))
print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[2], (memory_address + 4 * 2), 35, reg[t2], reg[t1], 0))
print("+%25s+%15s+%6s+%6s+%6s+%20s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 20))

# print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[3], hex_add(memory_address, 4 * 3), binf(5, 6), binf(reg[t1], 5), binf(reg[s5], 5), binf(bin(int(hex_add(memory_address, 4 * 7), 16))[2:], 16)))
print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[3], (memory_address + 4 * 3),5, reg[t1], reg[s5], 3))
print("+%25s+%15s+%6s+%6s+%6s+%20s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 20))

# print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[4], hex_add(memory_address, 4 * 4), binf(8, 6), binf(reg[s1], 5), binf(reg[s1], 5), binf(2, 16)))
print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[4], (memory_address + 4 * 4), 8, reg[s1], reg[s1], 2))
print("+%25s+%15s+%6s+%6s+%6s+%20s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 20))

# print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[5], hex_add(memory_address, 4 * 5), binf(8, 6), binf(reg[s1], 5), binf(reg[s1], 5), binf('1111111111111111', 16)))
print("|%25s|%15s|%6s|%6s|%6s|%20s|" % (instructions[5], (memory_address + 4 * 5), 8, reg[s1], reg[s1], -1))
print("+%25s+%15s+%6s+%6s+%6s+%20s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 6, '-' * 6, '-' * 20))

# print("|%25s|%15s|%6s|%34s|" % (instructions[6], hex_add(memory_address, 4 * 6), binf(2, 6), binf(int(hex_add(memory_address, 4 * 0), 16), 26)))
print("|%25s|%15s|%6s|%34s|" % (instructions[6], (memory_address + 4 * 6), 2, memory_address // 4))
print("+%25s+%15s+%6s+%34s+" % ('-' * 25, '-' * 15, '-' * 6, '-' * 34))

# print("|%25s|%15s|%41s|" % (instructions[7], hex_add(memory_address, 4 * 7), ''))
print("|%25s|%15s|%41s|" % (instructions[7], (memory_address + 4 * 7), ''))
print("+%25s+%15s+%41s+" % ('-' * 25, '-' * 15, '-' * 41))


print('\n\n')
print('=' * 80)
print('ANSWER FOR QUESTION 2')
print('=' * 80, end = '\n\n\n')

multiplicand = sum(id)
multiplier = -(multiplicand - 1)

print("Student ID          :", student_id)
print("Multiplicand, X     :", multiplicand)
print("Multiplier, -(X - 1):", multiplier)

result = multiply(binf(-multiplier, 10), binf(multiplicand, 20))

print("Two's complement:", twos_complement(-int(result, 2), 20), '(', -int(result, 2), ')')


print('\n\n')
print('=' * 80)
print('ANSWER FOR QUESTION 3')
print('=' * 80, end = '\n\n\n')


print('task02:')
print('     add $a0, $a0, $a1')
print('	    slti $t0, $a0,', sum(id))
print('	    beq $t0, $zero, else')
print('	    sub $s0, $a0, $a1')
print('	    j elseend')
print('	    else:')
print('		    sum $s0, $a1, $a0')
print('	    elseend:')
print('		    add $v0, $s0, $zero')
print('		    jr $ra')
```
