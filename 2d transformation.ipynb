```python
#translation
def ts(dx,dy):
  a=[[1,0,dx],
     [0,1,dy],
     [0,0,1]]
  return np.array(a)

#scale
def sc(sx,sy):
  a=[[sx,0,0],[0,sy,0],[0,0,1]]
  return np.array(a)

#rotation  
def rotate(theta):
  a=[[math.cos(math.radians(theta)),-math.sin(math.radians(theta)),0],
     [math.sin(math.radians(theta)),math.cos(math.radians(theta)),0],
     [0,0,1]]
  return np.array(a)

#reflectionx
def refx():
  a=[[1,0,0],[0,-1,0],[0,0,1]]
  return np.array(a)

def refy():
  a=[[-1,0,0],[0,1,0],[0,0,1]]
  return np.array(a)

#shearing
def shx(a):
 aa=[[1,a,0],[0,1,0],[0,0,1]]
 return np.array(aa)

def shy(a):
  aa=[[1,0,0],[a,1,0],[0,0,1]]
  return np.array(aa)

def shxy(a,b):
  aa=[[1,a,0],[b,1,0],[0,0,1]]
  return np.array(aa)

#def invm(a):
 # return a.transpose() 

def mat(aa):
  return np.array(aa)


def mdot(*a):
  c=a[0]
  for i in range(1,len(a)):
    c=np.dot(c,a[i])
  return c
  
  


```
