```
class person:
  def __init__(self,fname,lname):
    self.first=fname
    self.last=lname

  def returnName(self):
    return self.first+' '+self.last

  def __getitem__(self, k):
    if k == 'first':
      return self.first
    if k == 'lname':
      return self.last
  
  def __setitem__(self, k, v):
    if  k == 'first':
      self.first = v
    if k == 'lname':
      self.last = v

p1=person('sabrina','orin')
p1['first']='sab'
print(p1.returnName())
print(p1['lname'])
```
