```python
import math
hostNeeded=[2055,8700,256,700,458,1158,800,2,2,2,2,2,2,2,2,1,1,1]
hostNeeded.sort(reverse=True)
dataforoutput=['Host','Ips','Given','H.bit','N.bit']
output=[]
for i,j in enumerate(hostNeeded):
  output.append([j])
  output[i].append(j+2)
  output[i].append( 2**(math.ceil(math.log2(j+2))))
  output[i].append((math.ceil(math.log2(j+2))))
  output[i].append((32-math.ceil(math.log2(j+2))))
 
for i in dataforoutput:
  print(i,'\t',end='')
print()
for i in output:
  for j in i:
    print(j,'\t',end='')
  print()
  
```
