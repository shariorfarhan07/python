```python
def maxv(a,b):
  if a>b:
    return a
  return b
  
def minv(a,b):
  if a<b:
    return a
  return b
```

```python
class node:
  def __init__(self,val=""):
    self.value=val
    self.child=[]
  def addchild(self,a):
    self.child.append(a) 
```

```python
import random
random.seed(300)

noOfterm=1
branch=3
min=1
max=20
#noOfterm=int(input())
#branch=int(input())
#min=int(input())
#max=int(input())

def randomValue():
  global max
  global min
  return random.randint(min,max)


def makegraph(noOfterm,branch,n):
  if n==None:
    n=node()
  if noOfterm!=1:
    for i in range(branch):
      n.addchild(node())
      makegraph(noOfterm-1,branch,n.child[i])
  else:
    for i in range(branch):
      n.addchild(node(randomValue()))
  return n

a=makegraph(noOfterm*2,branch,None)


#print(a.child)
#print(a.child[0].child)
#print(a.child[1].child)
#print(a.child[2].child)
```

```python
count=0
import math
def minimax(node, depth, maximizing):
  global count
  if depth == 0:
    count+=1
    return node.value 
  
  if maximizing:
    maxEval = -math.inf
    for child in node.child:
      eval = minimax(child, depth - 1, False)
      maxEval = maxv(maxEval, eval)
    return maxEval
  else:
    minEval = math.inf
    for child in node.child:
      eval = minimax(child, depth - 1, True)
      minEval = minv(minEval, eval)
     
    return minEval
```

```python
import math
count=0
def minimaxAlphaBeta(node, depth,alpha, beta, maximizing):
  global count
  if depth == 0:
    count+=1
    return node.value 
  
  if maximizing:
    maxEval = -math.inf
    for child in node.child:
      eval = minimaxAlphaBeta(child, depth - 1,alpha, beta, False)
      maxEval = maxv(maxEval, eval)
      alpha = maxv(alpha, eval)
      if beta <= alpha:
        break
    return maxEval
  else:
    minEval = math.inf
    for child in node.child:
      eval = minimaxAlphaBeta(child, depth - 1,alpha, beta, True)
      minEval = minv(minEval, eval)
      beta = minv(beta, eval)
      if beta <= alpha:
        break     
    return minEval


#c=minimaxAlphaBeta(a, 2,-m.inf,m.inf, True)
#print(c)
#print(count)

#m=minimax(a, 2, True)
#print(count)
#print(m)

print("Depth:",noOfterm*2)

print("Branch:",branch)

print("Terminal States (Leaf Nodes):", branch**(noOfterm*2))
c=minimax(a, 2, True)
print("Maximum amount:", c)
print('Comparisons:', count)
c=minimaxAlphaBeta(a, noOfterm*2 ,float("-inf"),float("inf"), True)
print('Comparisons:', count)



```
