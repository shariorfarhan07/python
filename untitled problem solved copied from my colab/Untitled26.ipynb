```
n = int(input())
n1, n2 = 0, 1
count = 0
d={}
if n== 1:
  d[count]=n1
else:
  while count < n:
    d[count]=n1
    nth = n1 + n2
    n1 = n2
    n2 = nth
    count += 1

for i in range(3):
  s=int(input())-1
  if s>0 and s<n:
    print( d[s])
  else:
    print(print("No fibo"))
```

```
class Avengers: 
  count = 0

    
  def __init__(self,n,p):
    Avengers.count+=1
    self.name=n
    self.partner=p

  
  
  def super_powers(self,p1,p2): 
    self.pp1=p1
    self.pp2=p2
  
  def printAvengersDetail(self):
    print('Name:',self.name)
    print('Partner:',self.partner)
    print('Super powers:',self.pp1,',',self.pp2)








print('Total Avengers:', Avengers.count)
a1 = Avengers('Captain America', 'Bucky Barnes')
a1.super_powers('Stamina', 'Slowed ageing')
a2 = Avengers('Doctor Strange', 'Ancient One')
a2.super_powers('Mastery of magic', 'Gifted surgeon')
a3 = Avengers('Iron Man', 'War Machine')
a3.super_powers('Genius level intellect', 'Scientist ')
print("=========================")
a1.printAvengersDetail()
print("=========================")
a2.printAvengersDetail()
print("=========================")
a3.printAvengersDetail()
print("=========================")
print('Total Avengers:', Avengers.count)
```

```
class Laptop:
  def __init__(self, name):
    self.name = name
  def check(self):
    print('The Laptop is working properly')

class Asus(Laptop):
  def __init__(self, x):
    super().__init__(x) 
  def check(self):
    print('This is Asus',self.name)
    Laptop.check(self)

class Dell(Laptop):
  def __init__(self, x):
    super().__init__(x) 
  def check(self):
    print('This is Asus',self.name)
    Laptop.check(self)

    
a = Asus('Zenbook 14')
d = Dell('Inspiron')
a.check()
print('=========================')
d.check()
print('=========================')
```
