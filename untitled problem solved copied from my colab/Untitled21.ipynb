```
n=int(input("how many concerned word are there:"))
listofwords=[]
count=0
for i in range(n):
  count+=1
  words=input("Please enter censored/uncensored pair #"+str(count)+":").split()
  listofwords.append(words)




line=input('Please enter a sentence with lowercase letters only').split()
i=0
j=0
while i<len(listofwords):
  j=0
  while j<len(line):
    if listofwords[i][0]==line[j]:
      line[j]=listofwords[i][1]
    j=j+1  
  i=1+i
l=''
for i in line:
  l=l+" "+i
print("The censored sentence is as follows(Output):",l)




```
