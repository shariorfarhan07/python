```
dicthero= {'Breach' : 'Initiator' , 'Brim': 'Controller', 'Cypher': 'Sentinel', 
'Jett': 'Duelist', 'Omen': 'Controller', 'Phoenix': 'Duelist', 'Raze': 
'Duelist', 'Reyna': 'Duelist', 'Sage': 'Sentinel', 'Sova': 'Initiator', 
'Skye': 'Initiator' , 'Killjoy': 'Sentinel' }

#Now assume you will be playing a 
#match in Valorant but you and your friends need to pick a perfect team of 5 
#heroes based on their roles.  Perfect team consists of 2 duelists, 1 
#controller, 1 initiator and 1 sentinel
valid={'Duelist':2,'Controller':1,'Initiator':1,'Sentinel':1}
instring={'Duelist':0,'Controller':0,'Initiator':0,'Sentinel':0}

inputfromuser=input().split(' ')
for i in inputfromuser:
  instring[dicthero[i]]=instring[dicthero[i]]+1

def checkteam():
  for i in valid:
    if valid[i]<instring[i]:
      return False
  return True

if checkteam():
  print('We have a perfect team. Let’s win this game.')
else:
  print('We don’t have a perfect team. It will be hard to win this match.')
```
