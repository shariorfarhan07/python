```python
#Design a  team class so that the following output is produced upon 
#executing the following code:
#[𝐇𝐢𝐧𝐭: 𝐄𝐚𝐜𝐡 𝐭𝐞𝐚𝐦 𝐰𝐢𝐥𝐥 𝐡𝐚𝐯𝐞 𝐞𝐱𝐚𝐜𝐭𝐥𝐲 𝟑 𝐦𝐞𝐦𝐛𝐞𝐫𝐬]
class team:
  def __init__(self,a,b):
    self.name=a
    self.id=b
    self.member=[]

  def details(self):
    print("Team name: ",self.name,"\nTeam ID:",self.id)

  def add_member(self,*a):
    for i in a:
      self.member.append(i)
    print("Team ",self.name," has ",len(a)," members")
    print('Name of team members:',end='')
    
    s=""
    for i in range(len(self.member)):
      if i!=len(self.member)-1:
        s+=self.member[i]+","
      else:
        s+=self.member[i]+"."
    print(s)
    if len(a)<3:
      print("Status:  ",(3-len(a)),"  more member should be added")
    elif len(a)>3: 
      print("Status:  ",(len(a)-3)," member must be eliminated")
    else:
      print("Status:  ok")





    

#𝐂𝐨𝐝𝐞:
t1 = team("Team A","0001")
t2 = team("Team B","0002")
t3 = team("Team C","0003")


print('##########################')
print(t1.details())
print('##########################')
print(t2.details())
print('##########################')
print(t3.details())
print('##########################')
t1.add_member('Alice', 'Bob', 'Carol')
print('##########################')
t2.add_member('Dominic', 'Emory')
print('##########################')
t3.add_member('Frankie','Harper', 'Isabella', 'Josie')

```
