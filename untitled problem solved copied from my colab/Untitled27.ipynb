```
n=int(input())
d={'even':[],'odd':[]}
for i in range(n):
  n=int(input())
  if n%2==0:
    d['even']+=[n]
  else:
    d['odd']+=[n]
  
print(d)
```

```
d={'2':[],'3':[],'7':[],'other':[] }
for i in range(int(input())):
  n=int(input())
  if n%7==0:
    d['7']+=[n]
  elif n%3==0:
    d['3']+=[n]
  elif n%2==0:
    d['2']+=[n]
  else:
    d['other']+=[n]
  
print(d)
```
