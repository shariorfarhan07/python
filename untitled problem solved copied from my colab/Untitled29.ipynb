```
class Node:
  def __init__(self, v, n):
    self.val= v
    self.next = n

def properSubset(head1, head2):
  h1=head1
  h2=head2
  def checker(a,b):
    while a.next!=None:
      if a.val==b.val:
        return True
      a=a.next
    return False
  b=h2
  a=h1
  while b.next!=None:
    w=checker(a,b)
    if w:
      b=b.next
  return w


a1=Node(1,Node(2,Node(3,Node(4,Node(5,Node(6,Node(7,None)))))))  
b2=Node(1,Node(2,Node(3,Node(4,Node(5,None)))))
print(properSubset(a1,b2))
```
